```python
import pandas as pd

# Load the CSV files
train = pd.read_csv("semart_train.csv", encoding='latin-1', on_bad_lines='skip', engine='python', sep='\t')
val = pd.read_csv("semart_val.csv", encoding='latin-1', on_bad_lines='skip', engine='python', sep='\t')
test = pd.read_csv("semart_test.csv", encoding='latin-1', on_bad_lines='skip', engine='python', sep='\t')

# Concatenate them
full_df = pd.concat([train, val, test], ignore_index=True)
```

```python
print(full_df.shape)
print(full_df.columns)
print(full_df.isnull().sum())
```

```python
full_df.duplicated(subset=["IMAGE_FILE"]).sum()
```

```python
# Remove rows where AUTHOR contains variants of "unknown" or "ROMANESQUE PAINTER" (case-insensitive)
pattern = r"\b(?:unknown|anon|anonymous|romanesque\s+painter)\b"
full_df_clean = full_df[
    ~full_df["AUTHOR"].astype(str).str.contains(pattern, case=False, na=False)
].reset_index(drop=True)

print("Before:", len(full_df))
print("After :", len(full_df_clean))
```

```python
# Check if any UNKNOWN MASTER remains
full_df_clean[
    full_df_clean["AUTHOR"].str.contains("UNKNOWN MASTER", case=False, na=False)
]
```

```python
# Check if any UNKNOWN ICON PAINTER remains
full_df_clean[
    full_df_clean["AUTHOR"].str.contains("UNKNOWN ICON PAINTER", case=False, na=False)
]
```

```python
# Check if any ROMANESQUE PAINTER remains
full_df_clean[
    full_df_clean["AUTHOR"].str.contains("ROMANESQUE PAINTER", case=False, na=False)
]
```
