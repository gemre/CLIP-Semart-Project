```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("semart_full.csv", encoding='latin-1', on_bad_lines='skip', engine='python', sep=',')

plt.figure()
df['TIMEFRAME'].value_counts().sort_index().plot(kind='bar')
plt.title("Number of Artworks per Timeframe")
plt.xlabel("Timeframe")
plt.ylabel("Count")
plt.tight_layout()
plt.show()


plt.figure()
df['SCHOOL'].value_counts().plot(kind='bar')
plt.title("Number of Artworks per School")
plt.xlabel("School")
plt.ylabel("Count")
plt.tight_layout()
plt.show()


plt.figure()
df['TYPE'].value_counts().plot(kind='bar')
plt.title("Number of Artworks per Type")
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()


top_100_authors = df['AUTHOR'].value_counts().head(100)
top_100_authors


top_100_authors.plot(kind='bar', figsize=(15, 6))
plt.title("Number of Artworks for the Top 100 Authors")
plt.xlabel("Authors")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
