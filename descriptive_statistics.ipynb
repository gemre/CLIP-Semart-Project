```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("semart_train.csv", encoding='latin-1', on_bad_lines='skip', engine='python', sep='\t')

plt.figure()
df['TIMEFRAME'].value_counts().sort_index().plot(kind='bar')
plt.title("Number of Artworks per Timeframe")
plt.xlabel("Timeframe")
plt.ylabel("Count")
plt.tight_layout()
plt.show()


plt.figure()
df['SCHOOL'].value_counts().plot(kind='bar')
plt.title("Number of Artworks per School")
plt.xlabel("School")
plt.ylabel("Count")
plt.tight_layout()
plt.show()


plt.figure()
df['TYPE'].value_counts().plot(kind='bar')
plt.title("Number of Artworks per Type")
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
